# U.S. Medical Insurance Costs

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
df
```

```python
# average, Median, Max, and Min insurance cost :
mean = df['charges'].mean()
max_value = df['charges'].max()
min_value = df['charges'].min()
median_value = df['charges'].median()
print('The Highest insurance cost is '+ str(max_value) + ' dollars')
print('The Average insurance cost is '+ str(mean) + ' dollars')
print('The Median insurance cost is '+ str(median_value) + ' dollars')
print('The Lowest insurance cost is '+ str(mean) + ' dollars')
```

```python
# regions:
df['region'].value_counts()
```

```python
sns.countplot(x='region', data = df)
sns.set_style('whitegrid')
plt.ylabel('Number of Clients')
plt.xlabel('location')
plt.show()
```

```python
# Average Age:
average_age = df['age'].mean()
print('The Average age is ' + str(average_age) + ' years old. ')
```

```python
# Difference of the insurance cost between smokers and non_smokers:
smokers = df.loc[(df['smoker']=='yes').values]
smokers
```

```python
sns.lmplot(x="age", y="charges", hue="smoker", data = df, markers = ['o', 'v'])
sns.set_style('whitegrid')
plt.title('Smokers insurance cost')
plt.xlabel('Age')
plt.ylabel('Insurance cost')
plt.show()
```

```python
# Average insurance Cost for both genders:
```

```python
male_smoker_filter = smokers['sex'] == 'male'
male_smokers_cost = smokers.loc[male_smoker_filter]
male_smokers_cost
male_smokers_avg_cost = male_smokers_cost['charges'].mean()
print('The Average Insurance Cost for a male smoker is ' + str(male_smokers_avg_cost) + ' dollars')
```

```python
female_smoker_filter = smokers['sex'] == 'female'
female_smokers_cost = smokers.loc[female_smoker_filter]
female_smokers_cost
female_smokers_avg_cost = female_smokers_cost['charges'].mean()
print('The Average Insurance Cost for a female smoker is ' + str(female_smokers_avg_cost) + ' dollars')
```

```python
sns.barplot(x= 'sex', y= 'charges', data = smokers, color = 'salmon')
sns.set_style('whitegrid')
plt.title('Average Insurance Cost vs Gender ')
plt.ylabel('Insurance Cost')
plt.xlabel('Gender')
plt.show()
```
